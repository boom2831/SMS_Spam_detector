```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    alpha_numeric_char = [w for w in text if w.isalnum()]
    
    removed_stop_words = [w for w in alpha_numeric_char 
                          if w not in stopwords.words('english')]
    
    return removed_stop_words
```

```python
def text_normalization_ps(text):
    ps = PorterStemmer()
    stemmed = [ps.stem(w) for w in text]
    
    return " ".join(stemmed)
```

```python
def text_normalization_lm(text):
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(w) for w in text]
    
    return " ".join(lemmatized)
```

```python
nltk.download('wordnet')
```

```python
df['processed'] = df['input'].apply(transform_text)
```

```python
df['normalized_ps'] = df['processed'].apply(text_normalization_ps)
```

```python
df['normalized_lm'] = df['processed'].apply(text_normalization_lm)
```

```python
cv = CountVectorizer(analyzer = 'char', ngram_range = (3,4))
```

```python
X1 = cv.fit_transform(df['normalized_lm']).toarray()
```

```python
X1.shape
```

```python
X2 = cv.fit_transform(df['normalized_ps']).toarray()
X2.shape
```

```python
y = df['result'].values
```

```python
y.shape
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y, test_size = 0.2, random_state = 72)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size = 0.2, random_state = 34)
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
rfc = RandomForestClassifier()
```

```python
gnb.fit(X1_train, y1_train)
y1_pred1 = gnb.predict(X1_test)
print(accuracy_score(y1_test, y1_pred1))
print(confusion_matrix(y1_test, y1_pred1))
print(precision_score(y1_test, y1_pred1))
```

```python
gnb.fit(X2_train, y2_train)
y2_pred1 = gnb.predict(X2_test)
print(accuracy_score(y2_test, y2_pred1))
print(confusion_matrix(y2_test, y2_pred1))
print(precision_score(y2_test, y2_pred1))
```

***Lemmatizer has more accuracy with GaussianNB***
